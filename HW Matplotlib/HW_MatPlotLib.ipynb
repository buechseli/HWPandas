```python
unique_cities = cities.groupby(['city']).count().reset_index()
unique_cities
```

```python
unique_city_list = unique_cities.city.reset_index()
unique_city_list
```

```python
city_data_clean = pd.DataFrame({'type': cities.groupby('city').first()['type'],

                               'driver_count': cities.groupby('city').sum()['driver_count']}).reset_index()

city_data_clean
```

```python
rides_summarized = rides[['city','fare']].groupby(['city']).mean()
rides_summarized=rides_summarized.reset_index()
```

```python
rides_summarized1 = rides[['city','ride_id']].groupby(['city']).count()
rides_summarized1=rides_summarized1.reset_index()
rides_summarized1
```

```python
Data_Merged1 = pd.merge(rides_summarized1, rides_summarized , how='inner', on='city')
Data_Merged1
```

```python
Data_Merged = pd.merge(city_data_clean, Data_Merged1 , how='inner', on='city')
```
